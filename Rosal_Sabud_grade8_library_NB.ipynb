```python
!pip install -q firebase-admin requests

import requests
import firebase_admin
from firebase_admin import credentials, db
```

```python
# Load JSON from GitHub
url = "https://raw.githubusercontent.com/KarlTSabud/Rosal-Sabud-grade8-library-Repo/main/books.json"
data = requests.get(url).json()
```

```python
# Connect to Firebase
if not firebase_admin._apps:
    cred = credentials.Certificate("firebase_key.json")
    firebase_admin.initialize_app(cred, {
        'databaseURL': 'https://rosal-sabud-g8-library-db-default-rtdb.asia-southeast1.firebasedatabase.app/'
    })

# Upload books to Firebase
ref = db.reference("library")
ref.set(data)
```

```python
# Get books from Firebase
books = ref.child("books").get()

# Show all books
print("All Books:\n")
for book in books:
    print(book)
print()
```

```python
print("Book Titles:\n")
for book in books:
    print("-", book["title"])
print()
```

```python
print("Title & Available Copies:\n")
total_copies = 0
for book in books:
    title = book["title"]
    available = book["details"]["available"]
    copies = book["details"]["copies"]
    print(f"{title} - {available} available")
    total_copies += copies
print()

print(f"Total number of books: {total_copies}\n")
```
